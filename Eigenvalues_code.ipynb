```python

#simple eigenvalues from analytic solution

import numpy as np
import matplotlib.pyplot as plt

def energies(w, delt, J):
  E1 = w + J/4

  E2 = -J/4 + np.sqrt(J**2 + delt**2)/2

  E3 = -J/4 - np.sqrt(J**2 + delt**2)/2

  E4 = -w + J/4

  return E1, E2, E3, E4
#parameters
w = 1000
delt = 10
J=np.linspace(0,100,500)
E1, E2, E3, E4 = energies(w,delt,J)
plt.figure(figsize=(5,4))
plt.plot(J, E1, 'b', label='E1 (up,up state)')
plt.plot(J, E2, 'm',label='E2 (up,down state)')
plt.plot(J, E3, 'g-', label='E3 (down, up state)')
plt.plot(J, E4, 'r',label='E4 (down, down state)')
plt.grid()
plt.xlabel('Coupling constant J')
plt.ylabel('Energy E')
plt.xlim(0,100)
plt.legend(loc=1)
plt.show()
plt.figure(figsize=(5,4))
plt.plot(J, E2, 'm-', label='E2 (up,down state)')
plt.plot(J, E3, 'g-', label='E3 (down, up state)')
plt.grid()
plt.xlabel('Coupling constant J')
plt.ylabel('Energy E')
plt.xlim(0,100)
plt.legend()
```
